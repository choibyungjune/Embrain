```python
import pandas as pd 
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('[PPDB] data.csv')
```

```python
df
```

```python
dupl = df
```

```python
print(len(df.columns))
df.head()
```

```python
labels = []
for i in df.columns : 
    if i.startswith('C') or i.startswith('D') :
        labels.append(i)
```

```python
notlabel = []
for i in df.columns :
    if i not in labels :
        notlabel.append(i)
```

```python
notlabel
```

```python
notlabel.remove('id')
```

```python
dup = dupl.drop(columns = notlabel)
```

```python
dup = dup[0:100]
```

```python
dup.columns
```

```python
for i in dup.columns : 
    dup[i] = pd.to_numeric(dup[i], errors = 'ignore')
```

```python
df2 = pd.pivot_table(dup, index = 'id').fillna(0).astype(int)
```

```python
df2
```

```python
mat = cosine_similarity(df2)
```

```python
mat = pd.DataFrame(mat,index = list(df2.index), columns = list(df2.index))
```

```python
mat[0:10]
```

```python
df[df['id'].isin(['A0002320', 'A0008034'])]
```

```python
mat.values[[np.arange(mat.shape[0])]*2] = 0
mat.iloc[mat>=0.5,] = 1
mat.iloc[mat<0.5,] = 0
mat.astype(int) 
```

```python
mat.to_csv("association2.csv")
# 코사인 유사도 행렬
```
